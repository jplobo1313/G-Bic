```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
X = pd.read_csv("./dlblc.csv",index_col=0)
```

```python
plt.figure()
fig, ax1 = plt.subplots(figsize=(13,7),dpi=80,facecolor='w',edgecolor='k')

plt.axis('off')

sns.heatmap(data=X, cmap = "coolwarm")

fig.tight_layout()

#plt.show()

#plt.savefig("./Figure_1.pdf", dpi=1200, bbox_inches = 'tight')
#plt.close(1)
```

```python
f = list(np.array(X).flatten())
```

```python

fig = plt.figure(1)

sns.distplot(f, hist=True, kde=True, color = 'darkred', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
print(np.mean(f))
print(np.std(f))
print(np.min(f))
print(np.max(f))
```
