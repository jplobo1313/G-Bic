```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python

# Reading the Simulation
X2 = pd.read_csv("./Dataset3_data.tsv",index_col=0, sep = "\t")
```

```python
List = list(X2.drop(labels=["x10","x13","x39","x88"]).index)

NewList = List[0:50] + ["x10","x13","x39","x88"] + List[50:len(List)] 

X2 = X2.reindex(NewList)
```

```python
plt.figure()
fig, ax1 = plt.subplots(figsize=(13,7),dpi=80,facecolor='w',edgecolor='k')

sns.heatmap(data=X2, cmap = "coolwarm")


fontsize = 20

plt.xlabel('Time',fontsize=fontsize)
plt.xticks(rotation=90)

plt.ylabel('Observations',fontsize=fontsize)

l3 = ax1.legend(loc = 'center right', bbox_to_anchor=(1.1,0.5),ncol=1)
fig.tight_layout()

plt.savefig("./heatmap.pdf", dpi=1200, bbox_inches = 'tight')
#plt.close(1)
```

```python
X3 = X2.loc[["x10","x13","x39","x88"],["y57","y58","y59","y60","y61","y62","y63","y64","y65","y66","y67","y68","y69","y70","y71","y72","y73","y74"]]
```

```python
# Lets begin our figure...
fig, ax1 = plt.subplots(figsize=(13,7),dpi=80,facecolor='w',edgecolor='k')

sns.set(context='notebook', style='whitegrid')

sns.lineplot(data=X3.T, dashes = False)
sns.color_palette("colorblind")


fontsize = 20

plt.xlabel('Time',fontsize=fontsize)
plt.xticks(rotation=90)

l3 = ax1.legend(loc = 'center right', bbox_to_anchor=(1.1,0.5),ncol=1)
fig.tight_layout()

plt.savefig("./lineplot.pdf", dpi=1200, bbox_inches = 'tight')
```
