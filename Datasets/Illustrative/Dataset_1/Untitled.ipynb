```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
X = pd.read_csv("./Dataset1_data.tsv",index_col=0, sep = "\t")
```

```python
plt.figure()
fig, ax1 = plt.subplots(figsize=(13,7),dpi=80,facecolor='w',edgecolor='k')

sns.heatmap(data=X, cmap = "coolwarm")


fontsize = 20

plt.ylabel('Observations',fontsize=fontsize)
plt.xticks(rotation=90)

plt.xlabel('Attributes',fontsize=fontsize)

l3 = ax1.legend(loc = 'center right', bbox_to_anchor=(1.1,0.5),ncol=1)
fig.tight_layout()

#plt.show()

plt.savefig("./Figure_1.pdf", dpi=1200, bbox_inches = 'tight')
#plt.close(1)
```
